
# Generation, characterization, and visualization of density matrices.

In this notebook we see how to generate, cheracterize, and visualize some multi-qubit quantum states.  
First of all, we import the necessary dependencies. In this case, we just need the `density_matrix_tool` module.

```python
from density_matrix_tool import *
```

We now generate a random pure state. We use the `density_matrix_random` function, with arguments the dimension of the Hilbert space, the rank, and the number of vanishing diagonal elements.  
We can check if it really is a pure state. To this end, we can verify the following conditions: $\text{Tr} \left[ \rho^2 \right] = 1$, $S \left( \rho \right) = 0$, $\lvert \rho_{ij} \rvert^2 = \rho_{ii} \rho_{jj}$.

```python
rho_pure = density_matrix_random(dimension=8, rank=1, number_of_zeros = 0)
print('Purity:', purity(rho_pure))
print('Von Neumann entropy:', von_neumann_entropy(rho_pure))
for i in range(len(rho_pure)):
    for j in range(len(rho_pure)):
        if j > i:
            print('|rho_{}{}|^2='.format(i,j), np.round(np.square(np.abs(rho_pure[i,j])), decimals=5))
            print('rho_{}{}'.format(i,i),'*', 'rho_{}{}='.format(j,j), np.round(np.real(rho_pure[i,i]*rho_pure[j,j]),decimals=5))
            print('rho_{}{}'.format(i,i),'*', 'rho_{}{}'.format(j,j), '-', '|rho_{}{}|^2='.format(i,j), np.round(np.real(rho_pure[i,i]*rho_pure[j,j]),decimals=5) - np.round(np.square(np.abs(rho_pure[i,j])), decimals=5))
            print('###########')
```

We can do the same with a random mixed state, verifying the following conditions: $\text{Tr} \left[ \rho^2 \right] < 1$, $S \left( \rho \right) > 0$, $\lvert \rho_{ij} \rvert^2 < \rho_{ii} \rho_{jj}$

```python
rho_mixed = density_matrix_random(dimension=8, rank=7, number_of_zeros = 0)
print('Purity:', purity(rho_mixed))
print('Von Neumann entropy:', von_neumann_entropy(rho_mixed))
for i in range(len(rho_mixed)):
    for j in range(len(rho_mixed)):
        if j > i:
            print('|rho_{}{}|^2='.format(i,j), np.round(np.square(np.abs(rho_mixed[i,j])), decimals=5))
            print('rho_{}{}'.format(i,i),'*', 'rho_{}{}='.format(j,j), np.round(np.real(rho_mixed[i,i]*rho_mixed[j,j]),decimals=5))
            print('rho_{}{}'.format(i,i),'*', 'rho_{}{}'.format(j,j), '-', '|rho_{}{}|^2='.format(i,j), np.round(np.real(rho_mixed[i,i]*rho_mixed[j,j]),decimals=5) - np.round(np.square(np.abs(rho_mixed[i,j])), decimals=5))
            print('###########')
```

There are several ways to visualize the generated density matrices:

* `plot_density_matrix_3D()` represents the density matrix with a 3d barplot; the height of the bar is the modulus, while the phase is encoded in the colour of the columns, and shown in the colorbar. It accepts the density matrix as argument.
* `plot_density_matrix_2D()` represents the density matrix with two 2d plots, one for the modulus, and one for the phase. The values are encoded into the respective colorbars.

```python
plot_density_matrix_3D(rho_pure)
```

```python
plot_density_matrix_2D(rho_pure)
```

We can also generate some special multi-qubit states, such as *W* and *GHZ* states, respectively with the `density_matrix_W()` and the `density_matrix_GHZ()` methods. The argument for these functions is the number of qubits.

```python
Wstate = density_matrix_W(3)
plot_density_matrix_3D(Wstate)
```

```python
GHZstate = density_matrix_GHZ(5)
plot_density_matrix_3D(GHZstate)
```
